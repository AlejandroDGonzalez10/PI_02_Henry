```python
# Obtener la lista de símbolos de las empresas del índice S&P 500
sp500url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
data_table = pd.read_html(sp500url)
sp500_info = data_table[0]
tickers = sp500_info['Symbol'].tolist()

# Descargar los precios de cierre ajustados de todas las empresas
sp500_prices = yf.download(tickers, start="2000-01-01", end="2023-12-31")['Adj Close']

# Calcular los retornos diarios de todas las empresas
daily_returns = sp500_prices.pct_change()

# Calcular el retorno anualizado de cada empresa
annual_returns = ((1 + daily_returns.mean()) ** 252) - 1

# Obtener información adicional sobre cada empresa
sp500_info = sp500_info[['Symbol', 'Security', 'GICS Sector', 'GICS Sub-Industry' ]]
sp500_info.columns = ['Empresa', 'Nombre', 'Sector', 'Industria' ]
```

```python
# Unir los datos de precios y la información de las empresas
df = pd.concat([annual_returns, sp500_info.set_index('Empresa')], axis=1)
```

```python
df.to_csv('stockspy2.csv')
```
